```python
import requests
```

## 🌐 HTTP 요청 보내기

🔗 [requests 공식 문서](https://docs.python-requests.org/)

URL에 HTTP 요청을 보내고, 응답의 상태 코드와 성공 여부를 확인합니다.

[→ §200 강의노트](../notes/UD-04-200__http-and-requests.md)

```python
# 📌 §200 HTTP 요청
# 명언 사이트에 요청 보내기
response = requests.get("https://quotes.toscrape.com")

# 응답 확인
print(response.status_code)
# 예상 출력: 200

print(response.ok)
# 예상 출력: True
```

## 📦 응답 객체 확인

`response.text`로 서버가 보낸 HTML 원본을 확인합니다.

[→ §200 강의노트](../notes/UD-04-200__http-and-requests.md)

```python
# 📌 §200 응답 확인
# HTML 내용 확인 — 첫 200자만 출력
print(response.text[:200])
# 예상 출력:
# <!DOCTYPE html>
# <html lang="en">
# <head>
# 	<meta charset="UTF-8">
# 	<title>Quotes to Scrape</title>
# ...
```

## ⚠️ 상태 코드와 에러 처리

존재하지 않는 페이지에 요청하면 어떤 응답이 오는지 확인합니다.

[→ §200 강의노트](../notes/UD-04-200__http-and-requests.md)

```python
# 📌 §200 에러 처리
# 일부러 존재하지 않는 페이지에 요청
response_err = requests.get("https://httpbin.org/status/404")

print(response_err.status_code)
# 예상 출력: 404

print(response_err.ok)
# 예상 출력: False
```

## 🛡️ 실전 패턴: 조건 분기

`.ok`를 사용해 성공했을 때만 처리하는 안전한 패턴을 익힙니다.

[→ §200 강의노트](../notes/UD-04-200__http-and-requests.md)

```python
# 📌 §200 실전 패턴
# 실전 패턴: 성공했을 때만 처리
response = requests.get("https://quotes.toscrape.com")

if response.ok:
    html = response.text
    print(f"가져온 HTML 길이: {len(html)}자")
else:
    print(f"실패: {response.status_code}")
# 예상 출력: 가져온 HTML 길이: 11051자 (사이트 상태에 따라 다를 수 있음)
```

```python
r = requests.get("https://www.naver.com")
print(r.status_code, r.ok)
```

```python
headers = {"User-Agent": "Mozilla/5.0 (educational test)"}
r = requests.get("https://ko.wikipedia.org/wiki/Python", headers=headers)
print(r.status_code, len(r.text))
```
