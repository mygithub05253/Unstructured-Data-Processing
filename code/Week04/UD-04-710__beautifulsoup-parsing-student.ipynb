## 🔍 regex로 같은 데이터 추출하기

Week 03에서 배운 regex로도 HTML에서 데이터를 추출할 수 있습니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 regex — 실행만 하세요
#
# 먼저 HTML 원본을 보겠습니다 (한 명언의 태그):
#   <span class="text" itemprop="text">"The world as we have created..."</span>
#
#   itemprop 이란?
#   → class처럼 태그에 붙는 속성입니다.
#     구글 같은 검색엔진이 "이 내용이 무엇인지" 이해하도록 의미를 표시해둔 것.
#     스크래핑과는 무관하지만, HTML 소스에 그대로 등장하므로 regex가 이걸 포함해야 합니다.
#
# regex 패턴 풀이:
#   r'class="text" itemprop="text">(.*?)<'
#   ├─ class="text" itemprop="text">  ← 이 문자열이 있는 위치를 찾고
#   ├─ (.*?)                           ← 바로 뒤 내용을 캡처 (명언 텍스트)
#   └─ <                               ← 다음 태그 시작 < 직전까지만
#
# BS4 비교: soup.select('.text') — 같은 결과, 훨씬 짧음
# → itemprop 속성이 사라지거나 순서가 바뀌면 regex 패턴 전체를 고쳐야 합니다.
#   이것이 HTML에서 BS4가 더 안전한 이유입니다.

regex_quotes = re.findall(
    r'class="text" itemprop="text">(.*?)<', html
)
print(f"regex 결과: {len(regex_quotes)}개")
# 예상 출력: regex 결과: 10개
```
