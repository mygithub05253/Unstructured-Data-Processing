URL에 HTTP 요청을 보내고, 응답의 상태 코드와 성공 여부를 확인합니다.

[→ §200 강의노트](../notes/UD-04-200__http-and-requests.md)

```python
# 📌 §200 HTTP 요청
# ✍️ requests.get()으로 quotes.toscrape.com에 요청을 보내고 status_code와 ok를 확인해보세요
# 코드를 작성해주세요 👇
import requests

# 명언 사이트에 요청 보내기
response = requests.get("https://quotes.toscrape.com")

# 응답 확인
print(response.status_code)
# 예상 출력: 200

print(response.ok)
# 예상 출력: True

# 예상 출력: 200
# 예상 출력: True
```
