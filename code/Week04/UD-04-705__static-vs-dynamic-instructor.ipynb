# 🔄 UD-04-705 정적 vs 동적 페이지 비교

이 노트북에서는 `requests` 라이브러리를 사용해 정적/동적 페이지를 직접 비교하고, 주요 HTTP 응답 속성을 확인합니다.

```python
import requests
from bs4 import BeautifulSoup
```

## 1. 첫 번째 코드: 웹 페이지 가져오기 (UD-04-200)

가장 기본적인 `requests.get()` 사용법입니다.

```python
import requests

# 명언 사이트에 요청 보내기
response = requests.get("https://quotes.toscrape.com")

# 응답 확인
print(f"상태 코드: {response.status_code}")
print(f"성공 여부: {response.ok}")
```

```python
# HTML 내용 확인 — 첫 200자만 출력
print(response.text[:200])
```

## 2. 상태 코드 확인 (UD-04-200)

```python
# 일부러 존재하지 않는 페이지에 요청
response_err = requests.get("https://httpbin.org/status/404")

print(f"상태 코드: {response_err.status_code}")
print(f"성공 여부: {response_err.ok}")
```

## 3. 정적 vs 동적 비교 (UD-04-250)

```python
# 1) 정적 버전
r_static = requests.get("https://quotes.toscrape.com")
soup_static = BeautifulSoup(r_static.text, 'html.parser')
print(f"정적 페이지 명언 수: {len(soup_static.select('.text'))}")

# 2) 동적(JS) 버전
r_dynamic = requests.get("https://quotes.toscrape.com/js/")
soup_dynamic = BeautifulSoup(r_dynamic.text, 'html.parser')
print(f"동적 페이지 명언 수: {len(soup_dynamic.select('.text'))}")
```

## 🎯 [선택/심화] 다른 사이트 찔러보기 & POST 방식

여기서는 본문에서 다루지 않은 심화 사례(네이버, 위키백과, POST)를 실습합니다.

```python
# 네이버 메인 페이지
r = requests.get("https://www.naver.com")
print(f"네이버 응답: {r.status_code}, {r.ok}")

# 위키백과 (User-Agent 헤더 사용)
headers = {"User-Agent": "Mozilla/5.0 (educational test)"}
r = requests.get("https://ko.wikipedia.org/wiki/Python", headers=headers)
print(f"위키백과 응답: {r.status_code}, 길이: {len(r.text)}")
```

```python
# POST — 데이터 보내기
r_post = requests.post("https://httpbin.org/post", data={"username": "gachon", "message": "hello"})
print(f"POST 보낸 데이터: {r_post.json()['form']}")
```
