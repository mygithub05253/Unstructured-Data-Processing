```python
import requests
from bs4 import BeautifulSoup
import re

# §200에서 배운 코드 — 실행만 하세요
response = requests.get("https://quotes.toscrape.com")
html = response.text
```

## 🍲 1단계: 파싱 — HTML을 구조로 변환

🔗 [BeautifulSoup 공식 문서](https://www.crummy.com/software/BeautifulSoup/bs4/doc/)

HTML 문자열을 BeautifulSoup으로 탐색 가능한 트리 구조로 변환합니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 파싱
soup = BeautifulSoup(html, 'html.parser')
# soup 안에 HTML 전체가 트리 구조로 정리됨
```

## 🎯 2단계: 선택 — CSS 셀렉터로 요소 지정

`.select()` 메서드에 CSS 셀렉터를 넣어 원하는 요소를 리스트로 가져옵니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 선택
# 명언 텍스트 추출
quotes = soup.select('.text')
print(f"명언 수: {len(quotes)}")
# 예상 출력: 명언 수: 10

# 저자 이름 추출
authors = soup.select('.author')
print(f"저자 수: {len(authors)}")
# 예상 출력: 저자 수: 10
```

## 📝 3단계: 추출 — 텍스트 꺼내기

선택한 요소에서 `.text` 속성으로 순수 텍스트만 꺼냅니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 추출
# 첫 번째 명언과 저자
print(quotes[0].text)
# 예상 출력: “The world as we have created it is a process of our thinking.
#            It cannot be changed without changing our thinking.”

print(authors[0].text)
# 예상 출력: Albert Einstein
```

## 🔍 regex로 같은 데이터 추출하기

Week 03에서 배운 regex로도 HTML에서 데이터를 추출할 수 있습니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 regex
# regex로 명언 추출
regex_quotes = re.findall(
    r'class="text" itemprop="text">(.*?)<', html
)
print(f"regex 결과: {len(regex_quotes)}개")
# 예상 출력: regex 결과: 10개
```

## ⚖️ BeautifulSoup vs regex 비교

같은 데이터를 BeautifulSoup으로 추출하여 두 방법의 결과를 비교합니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 bs4 비교
# BeautifulSoup으로 명언 추출
bs4_quotes = [q.text for q in soup.select('.text')]
print(f"bs4 결과: {len(bs4_quotes)}개")
# 예상 출력: bs4 결과: 10개
```

## 🔗 속성 접근: 링크 추출

텍스트뿐 아니라 HTML 속성(href 등)도 `.get()`으로 추출할 수 있습니다.

[→ §300 강의노트](../notes/UD-04-300__beautifulsoup-devtools.md)

```python
# 📌 §300 속성
# 저자 상세 페이지 링크 추출
links = soup.select('.quote a')
for link in links[:3]:
    print(link.get('href'))
# 예상 출력:
# /author/Albert-Einstein
# /author/J-K-Rowling
# /author/Albert-Einstein
```
