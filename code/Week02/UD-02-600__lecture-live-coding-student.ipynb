## 🛠️ §200 파이썬 문자열 5대 메서드 + 함수화

→ [UD-02-200 강의노트](../notes/UD-02-200__string-five-methods.md)

```python
# 🧹 1. strip — 보이지 않는 공백 제거
raw = " Python "
# ✍️ strip()의 효과를 확인해보세요
# 코드를 작성해주세요 👇
print(raw == "Python")          
print(raw.strip() == "Python")  
```

```python
# 🔡 2. lower — 대소문자 통일
a = "Python"
b = "python"
# ✍️ lower()의 효과를 확인해보세요
# 코드를 작성해주세요 👇
print(a == b)
print(a.lower() == b)
```

```python
# 🔄 3a. replace — 기호·문자 치환 (숫자 변환)
price = "50,000원"
# ✍️ replace()를 체이닝하여 숫자만 추출해보세요
# 코드를 작성해주세요 👇
clean = price.replace(",", "").replace("원", "")
print(int(clean))
```

```python
# 🩹 3b. replace — 오타 교정
review = "맛잇다 또 올게요"
# ✍️ replace()를 사용해 오타를 교정해보세요
# 코드를 작성해주세요 👇
fixed = review.replace("맛잇다", "맛있다")
print(fixed)
```

```python
# ✂️ 4. split — 문자열을 리스트로 분리
row = "2026-03-01,apple,3000"
# ✍️ split(',')을 사용해 분리해보세요
# 코드를 작성해주세요 👇
row.split(",")
```

```python
# 👀 split() 숨겨진 동작 차이
# ✍️ 인자 유무에 따른 split()의 차이를 확인해보세요
# 코드를 작성해주세요 👇
row = "2026-03-01,apple,3000"
parts = row.split(",")
print(parts[1])
```

```python
# 🔗 5. join + split — 다중 공백 정규화
messy = "data    science    project"
# ✍️ split()과 join()을 연계하여 다중 공백을 정규화해보세요
# 코드를 작성해주세요 👇
clean = " ".join(messy.split())
print(clean)
```

```python
# 🧩 join() 이해: split()로 쪼개고 join()으로 이어붙임
text = "a b"
# ✍️ split() 후 다시 join()으로 합치는 과정을 실습해보세요
# 코드를 작성해주세요 👇
# split(): 텍스트(문자열)를 쪼갠다
pieces = text.split()
print(pieces)  # ['a', 'b']

# join(): 구분자(문자열)가 리스트를 이어붙인다
sep = " "
merged = sep.join(pieces)
print(merged)  # 'a b'
```

```python
# 📦 clean_text_basic() — 5개를 하나로 묶기
# ✍️ 정제 과정을 함수로 만들어보세요
# 코드를 작성해주세요 👇
def clean_text_basic(text: str) -> str:
    """문자열 기본 정제: lower → strip → 공백 정규화"""
    text = text.lower().strip()
    text = " ".join(text.split())
    return text
```

```python
# 🧪 3개 변형 → 모두 동일 출력
test_samples = ["  Data  Science  ", "DATA science", "data   science"]
# ✍️ 만든 함수를 테스트용 문구들에 적용해보세요
# 코드를 작성해주세요 👇
results = [clean_text_basic(s) for s in test_samples]
print(results)

# 예상 출력: ['data science', 'data science', 'data science']
```

```python
# ⛓️ 메서드 체이닝 — 한 줄에 여러 단계
# ✍️ 메서드 체이닝을 활용해 한 줄로 처리해보세요
# 코드를 작성해주세요 👇
result = " 맛잇다  또올게요 ".lower().strip().replace("맛잇다", "맛있다")
print(result)

# 예상 출력: 맛있다  또올게요
```

```python
# 🐢 for-loop 방식의 문제: NaN 방어 필수, 코드 5줄 이상
# ✍️ for 루프를 사용하여 DataFrame의 리뷰를 정제해보세요
# 코드를 작성해주세요 👇
for i in range(len(df)):
    if pd.notna(df.loc[i, "review"]):  # NaN 방어 필수
        df.loc[i, "clean"] = clean_text_basic(df.loc[i, "review"])
print(df)
```

```python
# ⛓️ Series.str 체인 데모 — 커머스 리뷰 정제
df_chain = pd.DataFrame({"review": [
    " 맛있어요 또 올게요!! ",
    "맛잇다 별로임",
    "  맛있다   추천합니다  ",
    None,
]})

# ✍️ .str 접근자를 체이닝하여 파이프라인을 구축해보세요
# 코드를 작성해주세요 👇
df_chain["clean"] = (
    df["review"]
    .str.lower()
    .str.strip()
    .str.replace("맛있다", "맛있다", regex=False)
)

print(df_chain[["review", "clean"]])
```

```python
# 📖 줄바꿈으로 파이프라인 가독성 확보
# ✍️ 가독성을 위해 괄호를 사용한 줄바꿈 체이닝을 작성해보세요
# 코드를 작성해주세요 👇
df["clean"] = (
    df["review"]
    .str.lower()      # 1단계
    .str.strip()      # 2단계
    .str.replace(...)  # 3단계
)
```

```python
# 🥊 str.replace() vs df.replace() — 부분 일치 vs 전체 일치
# ✍️ Series.str.replace와 DataFrame.replace의 차이를 확인해보세요
# 코드를 작성해주세요 👇
# str.replace: "맛잇" 부분만 매칭 → 치환
df["review"].str.replace("맛잇", "맛있")

# df.replace: "맛잇다 별로임" 전체가 일치해야 치환
df.replace({"review": {"맛잇다 별로임": "맛있다 별로임"}})
```

```python
# 🔍 str.contains() 맛보기
# ✍️ .str.contains()를 사용해 특정 단어가 포함된 리뷰를 찾아보세요
# 코드를 작성해주세요 👇
mask = df["review"].str.contains("맛있", na=False)
print(df[mask])

# 예상 출력: "맛잇다" 행은 제외, "맛있" 포함 행만 출력
```

```python
# 🛍️ 커머스 리뷰 활용형 폭발 관찰
reviews = ["맛있는 곳이에요", "맛있다 추천", "맛있어서 또 올게요"]
# ✍️ 커머스 리뷰 토큰들을 모아 집합(set)으로 만들고 고유 토큰 수를 확인해보세요
# 코드를 작성해주세요 👇
all_tokens = []
for r in reviews:
    all_tokens.extend(r.split())

print(f"고유 토큰: {len(set(all_tokens))}개")
print(set(all_tokens))

# 예상 출력: 고유 토큰: 7개
# {'맛있는', '곳이에요', '맛있다', '추천', '맛있어서', '또', '올게요'}
```
