## ✂️ 슬라이싱 — 범위로 자르기

슬라이싱은 `[start:stop]` 형태로 문자열의 일부를 추출합니다.  
- `start` 생략 → 처음부터  
- `stop` 생략 → 끝까지  
- `[::step]` → 간격을 지정

> **Q03** (11점)
>
> address에서 "강서구"를 슬라이싱으로 출력하세요.

```python
# ===코드 시작===
# address = "서울특별시 강서구 방화3동 827 국립국어원 National Institute of Korean Language"
# building = "국립국어원"
# district = "강서구"
print(address[6:9])

# ===코드 종료===
```

`step = -1`이면 역순으로 뒤집습니다. 회문(palindrome) 판별에 자주 사용됩니다.

> **Q04** (11점)
>
> address를 처음부터 끝까지 한 칸씩 건너뛰어(step=2) 출력하세요.

```python
# ===코드 시작===
print(address[::2])

# ===코드 종료===
```

## 🔍 type()과 f-string

`type()`은 값의 자료형을 확인합니다. 자료형에 따라 가능한 연산이 달라집니다.  
`f-string`은 변수 값을 문자열 안에 직접 삽입하는 방법입니다.

> **Q05** (11점)
>
> 정수 1과 실수 1.0의 타입을 각각 출력하세요.

```python
# ===코드 시작===
print(type(1))
print(type(1.0))

# ===코드 종료===
```

f-string은 `f"...{변수}..."` 형식으로, 문자와 숫자를 `str()` 변환 없이 이어붙입니다.

> **Q06** (11점)
>
> f-string으로 "{building}은 {district}에 있습니다."를 출력하세요.

```python
# ===코드 시작===
print(f"{building}은 {district}에 위치해 있습니다.")

# ===코드 종료===
```

## 🪤 `in` 연산자 — str vs list

> **Q07** (11점)
>
> "국어원"이 address 문자열에 포함되어 있는지 확인하세요.

```python
# ===코드 시작===
print("국어원" in address)

# ===코드 종료===
```

> **Q08** (11점)
>
> address를 단어 단위로 분리한 뒤, "국어원"이 목록에 있는지 확인하세요.

```python
# ===코드 시작===
address_list = address.split()
print(address_list)
print("국어원" in address_list)

# ===코드 종료===
```
