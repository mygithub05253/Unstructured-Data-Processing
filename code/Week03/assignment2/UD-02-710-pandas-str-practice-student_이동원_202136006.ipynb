```python
import pandas as pd

document = [
    "파이썬 기초 수업 신청 문의입니다.",
    " 데이터분석 프로젝트 참여 안내입니다.",
    "머신러닝 스터디 모집 문의입니다. ",
    "딥러닝 NLP 실습 안내입니다.",
    " NLP 입문 과목 신청 문의입니다."
]
df = pd.DataFrame(document, columns=["문서"])
print(df)
```

## 🔠 대소문자 변환

🔗 [pandas Series.str](https://pandas.pydata.org/docs/reference/series.html#api-series-str)

> **Q01** (12점)
>
> df["문서"] 열을 소문자로 변환하여 출력하세요.

```python
# ===코드 시작===
df_1 = df["문서"].str.lower()
print(df_1)

# ===코드 종료===
```

> **Q02** (12점)
>
> df["문서"] 열을 대문자로 변환하여 출력하세요.

```python
# ===코드 시작===
df_2 = df["문서"].str.upper()
print(df_2)

# ===코드 종료===
```

## 🧹 공백 제거

> **Q03** (12점)
>
> 앞뒤 공백을 제거한 새 열(문서_clean)을 만들고 원본 열과 나란히 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
print(df[["문서", "문서_clean"]])

# ===코드 종료===
```

## 🔢 문자 수와 단어 수

> **Q04** (12점)
>
> 문서_clean을 만들고 각 문서의 글자 수를 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
print(df["문서_clean"].str.len())

# ===코드 종료===
```

> **Q05** (13점)
>
> 문서_clean을 만들고 각 문서의 단어 수를 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
print(df["문서_clean"].str.split().str.len())

# ===코드 종료===
```

## 🔍 키워드 필터링

> **Q06** (13점)
>
> 문서_clean을 만들고 "안내"가 포함된 행을 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
print(df[df["문서_clean"].str.contains("안내")])
# ===코드 종료===
```

> **Q07** (13점)
>
> 문서_clean을 만들고 "NLP" 포함 여부를 True/False로 NLP여부 열에 추가하여, 문서_clean과 함께 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
df["NLP여부"] = df["문서_clean"].str.contains("NLP")
print(df[["문서_clean", "NLP여부"]])
# ===코드 종료===
```

## 🔄 문자열 바꾸기

> **Q08** (13점)
>
> 문서_clean을 만들고 "신청"→"접수", "모집"→"선발"로 교체한 문서_v3를 만들어, 문서_clean과 나란히 출력하세요.

```python
# ===코드 시작===
df["문서_clean"] = df["문서"].str.strip()
df["문서_v3"] = df["문서_clean"].str.replace("신청", "접수").str.replace("모집", "선발")
print(df[["문서_clean", "문서_v3"]])
# ===코드 종료===
```
